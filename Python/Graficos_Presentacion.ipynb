```python
import matplotlib.pyplot as plt
import numpy as np


labels = ['Central', 'Lateral']
ag12bic = [5.7230, 5.5288] #ag12bic
ag13 = [5.4429, 5.2582 ] #ag13

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, ag13, width, label='Ag13')
rects2 = ax.bar(x + width/2, ag12bic, width, label='Ag12Bic')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Magnetizaci√≥n [a0]')
ax.set_xticks(x)
ax.set_xticklabels(labels)
ax.legend()

fig.tight_layout()
plt.legend(loc='upper left', bbox_to_anchor=(1.05, 1),)
plt.show()
```
