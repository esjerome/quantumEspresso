```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use('seaborn-whitegrid')
import numpy as np
from matplotlib import rc
import pandas as pd
```

```python
from matplotlib.transforms import BboxBase
```

```python
x = np.array([15,20,25,30,35,40,45])
y_1 = np.array([-4618.321541,-4618.01606594,-4618.01106267,-4618.00978918,-4618.00973499,-4618.00974258,-4618.00999204])
y_2 = np.array([-5169.31605515,-5169.03862538,-5168.99786142,-5168.99545808,-5168.99542907,-5168.99524793,-5168.99523831])
y_3 = np.array([-5169.45991525,-5169.25042889,-5169.23521166,-5169.23374842,-5169.23372181, -5169.23355290,-5169.23353987])
```

```python
fig, axs = plt.subplots(2, 2, figsize=(9,9))

fig.subplots_adjust(left=3, bottom=4, right=4, top=5, wspace=0.4, hspace=0.4)

axs[0, 0].set_title('(A)',y=0, pad=-50, verticalalignment="top", fontsize=19)
axs[0, 0].set_xlabel('Longitud $[a_{0}]$', fontsize=15)
axs[0, 0].set_ylabel('Energa del Sistema [eV]',fontsize=15)
axs[0, 0].plot(x, y_1)
axs[0, 0].scatter(x, y_1)
listado = np.arange(y_1.min(), y_1.max()+(y_1.max()-y_1.min())/5, (y_1.max()-y_1.min())/5)[:6]
axs[0, 0].set_yticks(listado)
axs[0, 0].set_yticklabels([str(round(r,2)) for r in listado])

#Cluster de Ag12BiC
axs[0, 1].set_title('(B)',y=0, pad=-50, verticalalignment="top", fontsize=19)
axs[0, 1].set_xlabel('Longitud $[a_{0}]$', fontsize=15)
axs[0, 1].set_ylabel('Energa del Sistema [eV]',fontsize=15)
axs[0, 1].plot(x, y_2)
axs[0, 1].scatter(x, y_2)
listado = np.arange(y_2.min(), y_2.max()+(y_2.max()-y_2.min())/5, (y_2.max()-y_2.min())/5)[:6]
axs[0, 1].set_yticks(listado)
axs[0, 1].set_yticklabels([str(round(r,2)) for r in listado])


#Cluster de Ag12BiL
axs[1, 0].set_title('(C)',y=0, pad=-50, verticalalignment="top", fontsize=19)
axs[1, 0].set_xlabel('Longitud $[a_{0}]$', fontsize=15)
axs[1, 0].set_ylabel('Energa del Sistema [eV]',fontsize=15)
axs[1, 0].plot(x, y_3)
axs[1, 0].scatter(x, y_3)
listado = np.arange(y_3.min(), y_3.max()+(y_3.max()-y_3.min())/5, (y_3.max()-y_3.min())/5)[:6]
axs[1, 0].set_yticks(listado)
axs[1, 0].set_yticklabels([str(round(r,2)) for r in listado])


fig.delaxes(axs[1][1])
fig.savefig('../../Tesis/Anexos/imagenes/nuevas_convergencias/celda.png', format='png',bbox_inches='tight')


plt.show()
```
